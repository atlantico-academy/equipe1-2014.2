<div id='contextualizacao'><h3>1.1. Contextualização</h3></div> 

O conjunto de dados **Sleep Efficiency** contém informações sobre um grupo de indivíduos testados e seus padrões de sono. Cada indivíduo é identificado por um único `ID`, e sua idade e gênero também são registrados. 

As variáveis `Bedtime` e `Wakeup time` indicam, respectivamente, quando cada indivíduo vai para a cama e acorda todos os dias. 

A variável `Sleep Duration` registra a quantidade total de tempo que cada indivíduo dormiu, em horas.

A variável `Sleep efficiency` é definida como uma medida da proporção de tempo gasto na cama que é passado efetivamente dormindo. 

As variáveis `Light sleep percentage`, `Deep sleep percentage` e `REM sleep percentage` indicam o percentual de tempo que cada indivíduo gastou em cada estágio do sono. 

A variável `Awakenings` registra o número de vezes que cada sujeito acorda durante a noite. 

Além disso, o conjunto de dados inclui informações sobre o consumo de cafeína e álcool de cada indivíduo nas 24 horas anteriores à hora de dormir, seu status de consumo de tabaco e sua frequência semanal na prática de exercícios físicos.

---

**Sono leve (Light Sleep)**

O sono leve é a fase inicial do ciclo do sono, representando uma transição entre a vigília e o sono profundo. Durante essa fase, que ocupa cerca de 50% do tempo total de sono, o corpo começa a relaxar, a respiração e os batimentos cardíacos desaceleram, e a temperatura corporal diminui. O sono leve é dividido em dois estágios: o primeiro estágio, onde se pode ser facilmente acordado, e o segundo estágio, onde o corpo se prepara para o sono profundo. O sono leve é importante para o ajuste gradual do corpo a estados mais profundos de descanso. Essa fase também desempenha um papel fundamental em manter a continuidade do ciclo do sono, facilitando a transição entre as diferentes fases do sono.

---

**Sono profundo (Deep Sleep)**

O sono profundo, também conhecido como sono de ondas lentas, é a fase mais restauradora do sono, na qual o corpo realiza a maioria de seus processos regenerativos. Durante esta fase, que ocorre mais frequentemente na primeira metade da noite, a atividade cerebral é caracterizada por ondas cerebrais lentas e de alta amplitude. O corpo concentra-se na reparação de músculos e tecidos, no fortalecimento do sistema imunológico, e na liberação de hormônios de crescimento essenciais para o desenvolvimento físico. O sono profundo é vital para a recuperação física, o crescimento, e a saúde metabólica. A interrupção desta fase pode resultar em sensação de cansaço, dificuldade de concentração e comprometimento do sistema imunológico.

**Sono REM (REM Sleep)**

O sono REM, ou Rapid Eye Movement, é uma fase única do ciclo do sono caracterizada por movimentos oculares rápidos, aumento da atividade cerebral, sonhos vívidos, e uma quase paralisia muscular que impede que o corpo reaja fisicamente aos sonhos. Esta fase ocorre aproximadamente 90 minutos após o início do sono e se repete várias vezes durante a noite, com os períodos de REM aumentando em duração a cada ciclo. Durante o sono REM, o cérebro processa informações, consolida memórias, e estabiliza emoções, desempenhando um papel crítico no aprendizado e na adaptação emocional. A privação do sono REM pode levar a problemas cognitivos e emocionais, ressaltando sua importância para a saúde mental e o bem-estar geral.

<div id='outliers'><h3>1.6. Dados discrepantes (Outliers)</h3></div>

```python
# Identificação de outliers usando o método do Intervalo Interquantil (IQR)
numeric_df = df.select_dtypes(include=[np.number])
factor = 1.5

def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - factor * IQR
    upper_bound = Q3 + factor * IQR
    return (data < lower_bound) | (data > upper_bound)

# Aplicando a função a todas as colunas numéricas
outliers = numeric_df.apply(detect_outliers_iqr)

# Filtrando os dados para mostrar apenas os outliers
outliers_summary = outliers.sum().sort_values(ascending=False)
outliers_summary = outliers_summary[outliers_summary > 0]
outliers_data = numeric_df.loc[:, outliers_summary.index]
display(Markdown("*Número de Outliers por Coluna:* \n"))
display(outliers_summary)
print("\n")
display(outliers_data.head())
print("\n")

# Visualização gráfica
num_cols = outliers_data.shape[1]
num_rows = (num_cols + 1) // 2  
fig = plt.figure(figsize=(14, num_rows * 4))

fig.suptitle('Distribuição de dados discrepantes', fontweight='bold', fontsize=16, y=1)

for i, column in enumerate(outliers_data.columns, 1):
    ax = plt.subplot(num_rows, 2, i)
    sns.histplot(numeric_df[column], bins=30, kde=True, color=colors[3], label='Dados')
    sns.histplot(numeric_df[column][outliers[column]], bins=30, kde=True, color=colors[1], label='Outliers', alpha=0.7)
    ax.set_title(f'Distribuição de {column}', fontsize=12, fontweight='bold')
    ax.set_xlabel(column)
    ax.set_ylabel('Frequência')
    ax.legend()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

* Segundo o cálculo do IQR, foi observada a presença de valores discrepantes (outliers) nas variáveis `Sleep duration `, `Deep sleep percentage `, `Light sleep percentage ` e `Caffeine consumption`.

* A variável `Deep sleep percentage` possui o maior número de valores discrepantes, **60 outliers ao todo**, seguido de `Sleep duration`, com **16 outliers**, `Caffeine consumption`, com **4 outliers**, e `Light sleep percentage` com **2 outliers**.


<div id='padronizacao'><h3>1.7. Padronização de Variáveis</h3></div>

A partir da verificação dos tipos das variáveis, foi realizada a conversão dos tipos object das variáveis `Gender` e `Smoking Status` para dado categórico para otimizar as análises estatíticas. 

```python
#Convertendo a variáveis Gender e Smoking_status para o tipo categórico
df['Gender'] = df['Gender'].astype('category')
df['Smoking status'] = df['Smoking status'].astype('category')

print("Tipos dos dados após conversão: /n")
print(df.dtypes)
```

O próximo passo foi a separação da conversão das colunas `Bedtime` e `Wakeup Time` para **datetime**, assim como a separação dos dados de data e hora em quatro colunas distintas: `Bedtime_date`, `Bedtime_hour`, `Wakeup_data` e `Wakeup_hour`.

```python
#Convertendo a variáveis Bedtime e Wakeup Time para datetime
df['Bedtime'] = pd.to_datetime(df['Bedtime'])
df['Wakeup time'] = pd.to_datetime(df['Wakeup time'])

# Separando em data e hora para 'Bedtime'
df['bedtime_date'] = df['Bedtime'].dt.date
df['bedtime_hour'] = df['Bedtime'].dt.time

# Separarando em data e hora para 'Wakeup time'
df['wakeup_date'] = df['Wakeup time'].dt.date
df['wakeup_hour'] = df['Wakeup time'].dt.time

df = df.drop(columns=['Bedtime', 'Wakeup time'])

display(df.head())
```

Em seguida, foi realizada a conversão da unidade **onça (oz)** para **gramas (g)** para padronização e precisão das unidades de medida. A onça é uma unidade de peso do sistema imperial dos EUA. 1 onça equivale a 28,3495 gramas.

```python
# Conversão da unidade onça(oz) para gramas(g)
def oz_to_g(oz):
    return oz * 28.3495

# Aplicando a conversão na coluna 'Alcohol consumption'
df['alcohol_consumption_g'] = df['Alcohol consumption'].apply(oz_to_g)
df = df.drop(columns=['Alcohol consumption'])

display(df.head().round(2))
```

Por fim, as variáveis que continham espaços ou letras maiúsculas foram formatadas para o padrão **snake_case** do Python para evitar eventuais erros de análise. Esse padrão é caracterizado pelo uso de letras minúsculas e underscores.

```python
# Renomeando colunas
df.rename(columns = {'ID': 'id', 'Age': 'age', 'Gender': 'gender', 'Wakeup time':'wakeup_time', 'Sleep duration':'sleep_duration_hour',"Sleep efficiency":"sleep_efficiency",
                     "REM sleep percentage":"rem_sleep_percentage","Deep sleep percentage":"deep_sleep_percentage",
                     "Light sleep percentage":"light_sleep_percentage","Caffeine consumption":"caffeine_consumption_mg",
                     "Smoking status":"smoking_status","Exercise frequency":"exercise_frequency", "Awakenings":"awakenings"},
          inplace = True)
```

<div id='pergunta_16'><h3>3.16. Qual é a relação entre o horário de dormir, a duração do sono e a eficiência do sono?</h3></div>

>***Hipótese:*** *Indivíduos que dormem mais cedo e têm uma maior duração do sono tendem a ter maior eficiência do sono.*

```python
# Qual é a relação entre o horário de dormir, a duração do sono e a eficiência do sono?

df = pd.DataFrame({
    'wakeup_hour': ['22:00', '23:00', '00:00', '01:00', '22:00', '23:00', '00:00', '01:00'],
    'sleep_duration_hour': [8, 7, 6, 5, 8, 7, 6, 5],
    'sleep_efficiency': [85, 80, 75, 70, 90, 85, 80, 75]
})

plt.figure(figsize=(10, 6))
ax = sns.scatterplot(data=df, 
            x='wakeup_hour', 
            y='sleep_efficiency', 
            hue='sleep_duration_hour', 
            palette=colors, 
            size='sleep_duration_hour', 
            sizes=(50, 200))

plt.title('Relação entre Horário de Dormir, Duração do Sono e Eficiência do Sono',
          fontweight='bold',
          fontsize=14,
          y=1.02)

plt.xlabel('Horário de Dormir')
plt.ylabel('Eficiência do Sono (%)')
plt.legend(title='Duração do Sono (Horas)')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

* Há uma boa eficiência do sono (cerca de 85%) ao dormir às **22:00h com 7 horas de sono**.

* A eficiência do sono tende a diminuir gradualmente conforme o horário de dormir avança (23:00h, 00:00h e 01:00h).

* Dormir às **00:00h ou mais tarde** resulta em uma eficiência de sono inferior a **80%**.

* Dormir antes das **23:00h** parece estar associado a uma eficiência de sono superior a **85%**.

* **7 e 8 horas de sono** tendem a estar associadas a uma eficiência de sono superior, independentemente do horário de dormir.

* Apenas **5 horas de sono** estão consistentemente associadas a uma eficiência de sono abaixo de **80%**, independentemente do horário de dormir.
